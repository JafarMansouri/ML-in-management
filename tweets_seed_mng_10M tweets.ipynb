```python
import re
from io import StringIO
import pickle
import html
from html.parser import HTMLParser
```

```python
class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```python
with open("./seed_tweets_mng10m.txt", "rb") as fp:   
    seed_tweets_mng = pickle.load(fp)
```

```python
print('Number of seed tweets: ' , len(seed_tweets_mng))
```

# First 5000 non repetitive tweets after some cleanings

```python
for i in seed_tweets_mng[0:5000]:
    temp = strip_tags(i)
    #temp = re.sub("|", '''"''', temp)  
    #temp = re.sub("|||`", "'", temp)  
    temp = re.sub(r'[^\x00-\x7F]+|,',' ', temp)  
    temp = temp.lower() 
    temp = re.sub(r'(http://|https://)\S+', '', temp) 
    temp = re.sub(r'www\.\S+', '', temp)
    temp = re.sub('( |^)@\S+', '', temp) 
    #temp= re.sub("[\"\+\-\|\*\?\(\)\/\\\^\[\]``<>\.{}`'_;,@:~!\=%&]+", ' ',temp) 
    temp = temp.encode('utf8').decode('ISO-8859-1') # decode('ISO-8859-1') or decode('latin-1') or decode('windows-1252')
    print(temp)
    print('-------------')
```
