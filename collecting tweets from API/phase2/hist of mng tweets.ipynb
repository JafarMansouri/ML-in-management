```python
import psycopg2
con = psycopg2.connect(database="postgres", user="postgres", password="Jafarsql", host="localhost", port="5432")
print("Database opened successfully")

cur = con.cursor()
cur.execute("SELECT tweet_created_at from unique_tweets_manager_1")
rows = cur.fetchall()
# there is no repetative tweet
```

```python
d = dict()
for row in rows:
    info=row[0].split('-')
    if str(info[0])+'-'+str(info[1]) not in d:
        d[ str(info[0])+'-'+str(info[1]) ] = 1 
    else:
        d[ str(info[0])+'-'+str(info[1])  ] +=  1    
            
print("Operation done successfully")
con.close()
```

```python
d_new=dict(sorted(d.items()))
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})

plt.bar(d_new.keys(), d_new.values(),color='g')
plt.xlabel('years and months')
plt.ylabel('number of tweets')
plt.savefig("histogram.jpg")
plt.savefig('histogram.pdf')
plt.show()
```
