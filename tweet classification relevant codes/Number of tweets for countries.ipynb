```python
import sys
import pandas as pd
from geotext import GeoText
import time
import operator
```

```python
df_ent = pd.read_csv('ent_tweets_world.csv', delimiter='\t', na_values=".",error_bad_lines=False)#,warn_bad_lines=False)
print(df_ent.shape)
print(df_ent.columns)

rows_ent=list(df_ent[['user_id', 'location_profile']].itertuples(index=False, name=None)) #rows_ent0
del df_ent
print("Number of tweets in ent:",len(rows_ent))  #rows_ent0
print('Memory size of ent:',sys.getsizeof(rows_ent))
```

```python
df_mng = pd.read_csv('mng_tweets_world.csv', delimiter='\t', na_values=".",error_bad_lines=False,warn_bad_lines=False)
print(df_mng.shape)
print(df_mng.columns)

rows_mng=list(df_mng[['user_id', 'location_profile']].itertuples(index=False, name=None)) #rows_mng0

del df_mng
print(len(rows_mng)) #rows_mng0
print('size of mng:', sys.getsizeof(rows_mng)) #rows_mng0
```

```python
US_states =set(["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"])
```

```python
def find_country(text):
    if text == text:
    
        country_temp=set()
        
        if ("US" in text) or ("USA" in text) or ('Florida' in text):
            country_temp.add('US')
            
        elif "UK" in text:
            country_temp.add('GB')
            
        elif ("Sydney" in text) or   ("Melbourne" in text):
            country_temp.add('AU')
            
        else:
            sub_text=text.split(' ')
            for sub in sub_text:
                if sub in US_states:
                    country_temp.add('US')
                
        if len(country_temp)==0:
            countries = GeoText(text).country_mentions
            country_temp.update(set(list(countries))) 
            
        if len(country_temp)==0:
            country_temp=set('0')
        
    else:
        country_temp=set('0')
            
    country_temp = ', '.join(country_temp)

    return country_temp
```

```python
t0=time.time()

dic_countries_ent=dict()

for text in rows_ent:
    temp=find_country(text[1])      
    
    if temp not in dic_countries_ent:
        dic_countries_ent[temp]= 1
    else:
        dic_countries_ent[temp] +=1
        
print(time.time()-t0)        
```

# Number of tweets for ent (based on detected countries)

```python
dict(sorted( dic_countries_ent.items(),key=operator.itemgetter(1), reverse=True))
```

```python
t0=time.time()

dic_countries_mng=dict()

for text in rows_mng:
    temp=find_country(text[1])
    
    if temp not in dic_countries_mng:
        dic_countries_mng[temp]= 1
    else:
        dic_countries_mng[temp] +=1
        
print(time.time()-t0)        
```
