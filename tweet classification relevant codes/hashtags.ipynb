As it can be seen from the results, many hashtags have few repetitions. Here, a repetition is the number of the appearance of a hashtag for all people in the dataset (including all appearances of that hashtag for one person, for example, one person may use #Brazil some times.) 

About one sixth of the hashtags in ent and mng are common in the two sets. 
In general, working on hashtags does not seem very useful since many of them have few repetitions and also many of them seem irrelevant. 

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
import torch
import matplotlib.pyplot as plt
import pickle
import time
import numpy as np
import time 
import sys
# OPTIONAL: if you want to have more information on what's happening, activate the logger as follows
import logging
#logging.basicConfig(level=logging.INFO)
```

```python
#accessing to tweets 
with open("tweet_ent.txt", "rb") as fp:   # Unpickling
  tweet_ent = pickle.load(fp)
  
with open("tweet_mng.txt", "rb") as fp:   # Unpickling
  tweet_mng = pickle.load(fp)
```

```python
def cleaning (text):

    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    #removing emoji
    text = emoji_pattern.sub(r' ', text)   
    
    #replace consecutive non-ASCII characters with a space
    text = re.sub(r'[^\x00-\x7F]+|,Ä¶',' ', text)
    
    #removing hashtag sign
    text=re.sub('#', '', text) 
    
    text= re.sub("[\"\“\”\+\-\|\*\?\(\)\/\\\^\[\]\.{}`;•«,@:~!=%&]+", ' ',text) # except ' ’ _
    
#    text=re.sub(' +',' ',text)
    
    text=''.join(text)
   
    return text
```

```python
ent_hashtags_num=0
ent_hashtags_tweet=[]
for tweet in tweet_ent:
    ans=[]
    for  i in tweet.split():
        if i.startswith("#"):
            ans.extend(cleaning(i).split())
    if ans!=[]:
#        ans=' '.join(ans)
#        ent_hash_tweet.append(ans)
        ent_hashtags_tweet.extend(ans)
        ent_hashtags_num +=1
```

```python
mng_hashtags_num=0
mng_hashtags_tweet=[]
for tweet in tweet_mng:
    ans=[]
    for  i in tweet.split():
        if i.startswith("#"):
            ans.extend(cleaning(i).split())
    if ans!=[]:
#        ans=' '.join(ans)
#        mng_hash_tweet.append(ans)
        mng_hashtags_tweet.extend(ans)
        mng_hashtags_num +=1
```

```python
mng_hashtags_all=set(mng_hashtags_tweet)
ent_hashtags_all=set(ent_hashtags_tweet)
```

```python
print("number of total hashtags in ent:",len(ent_hashtags_tweet)) 
print("number of total hashtags in mng:",len(mng_hashtags_tweet)) 

print("number of non repeated hashtags in ent:",len(ent_hashtags_all))
print("number of non repeated hashtags in mng:",len(mng_hashtags_all))
```

```python
mng_hashtags_unique=mng_hashtags_all-ent_hashtags_all
ent_hashtags_unique=ent_hashtags_all-mng_hashtags_all
mng_ent_hashtags_common=mng_hashtags_all & ent_hashtags_all
```

```python
print("number of unique hashtags in ent:",len(ent_hashtags_unique))
print("number of unique hashtags in mng:",len(mng_hashtags_unique))
print("number of common hashtags in ent and mng:",len(mng_ent_hashtags_common))
```

```python
d_mng_all = dict()
for hashtag in mng_hashtags_tweet:
    if hashtag not in d_mng_all:
        d_mng_all[ hashtag ] = 1 
    else:
        d_mng_all[ hashtag ] +=1  
```

```python
d_ent_all = dict()
for hashtag in ent_hashtags_tweet:
    if hashtag not in d_ent_all:
        d_ent_all[ hashtag ] = 1 
    else:
        d_ent_all[ hashtag ] +=1  
```
